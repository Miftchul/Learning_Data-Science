# Hari 2: Jenis-Jenis Data dan Cara Mengumpulkannya

## Import Library yang Dibutuhkan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur tampilan grafik
plt.style.use('seaborn-v0_8')
sns.set(font_scale=1.2)

# Mengabaikan warning
import warnings
warnings.filterwarnings('ignore')
```

## 1. Jenis-Jenis Data

### 1.1 Data Numerik

```python
# Contoh data numerik
# 1. Data Diskrit (bilangan bulat)
jumlah_anak = [0, 1, 2, 3, 4, 2, 1, 0, 2, 3]

# 2. Data Kontinu (bilangan desimal)
tinggi_badan = [165.5, 170.2, 168.7, 172.3, 175.0, 163.8, 169.4, 182.5, 167.6, 178.9]

# Membuat DataFrame
data_numerik = pd.DataFrame({
    'Jumlah Anak': jumlah_anak,
    'Tinggi Badan (cm)': tinggi_badan
})

print("Data Numerik:")
display(data_numerik)

# Statistik deskriptif
print("Statistik Deskriptif:")
display(data_numerik.describe())
```

### 1.2 Data Kategorikal

```python
# Contoh data kategorikal
# 1. Data Nominal (tidak ada urutan)
jenis_kelamin = ['Laki-laki', 'Perempuan', 'Laki-laki', 'Laki-laki', 'Perempuan', 
                'Perempuan', 'Laki-laki', 'Laki-laki', 'Perempuan', 'Laki-laki']

# 2. Data Ordinal (ada urutan)
pendidikan = ['SMA', 'S1', 'S2', 'SMP', 'S1', 'SMA', 'S3', 'S1', 'SMA', 'S2']

# Membuat DataFrame
data_kategorikal = pd.DataFrame({
    'Jenis Kelamin': jenis_kelamin,
    'Pendidikan': pendidikan
})

print("Data Kategorikal:")
display(data_kategorikal)

# Frekuensi data kategorikal
print("Frekuensi Jenis Kelamin:")
display(data_kategorikal['Jenis Kelamin'].value_counts())

print("Frekuensi Pendidikan:")
display(data_kategorikal['Pendidikan'].value_counts())
```

### 1.3 Data Waktu (Time Series)

```python
# Contoh data waktu (time series)
tanggal = pd.date_range(start='2023-01-01', periods=10, freq='D')
suhu = [28, 29, 30, 27, 26, 28, 31, 30, 29, 28]

# Membuat DataFrame
data_waktu = pd.DataFrame({
    'Tanggal': tanggal,
    'Suhu (°C)': suhu
})

print("Data Time Series:")
display(data_waktu)

# Plot data time series
plt.figure(figsize=(10, 5))
plt.plot(data_waktu['Tanggal'], data_waktu['Suhu (°C)'], marker='o', linestyle='-')
plt.title('Perubahan Suhu Harian')
plt.xlabel('Tanggal')
plt.ylabel('Suhu (°C)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 2. Cara Mengumpulkan Data

### 2.1 Membaca Data dari File CSV

```python
# Membuat contoh data untuk disimpan sebagai CSV
data = {
    'Nama': ['Andi', 'Budi', 'Citra', 'Deni', 'Eka'],
    'Usia': [25, 30, 22, 35, 28],
    'Gaji': [5000000, 7500000, 4500000, 9000000, 6500000],
    'Departemen': ['IT', 'Marketing', 'Finance', 'HR', 'IT']
}

df = pd.DataFrame(data)

# Menyimpan data ke CSV
df.to_csv('data_karyawan.csv', index=False)
print("Data telah disimpan ke file CSV")

# Membaca data dari CSV
df_baca = pd.read_csv('data_karyawan.csv')
print("Data yang dibaca dari file CSV:")
display(df_baca)
```

### 2.2 Membaca Data dari Excel

```python
# Menyimpan data ke Excel
df.to_excel('data_karyawan.xlsx', index=False)
print("Data telah disimpan ke file Excel")

# Membaca data dari Excel
try:
    df_excel = pd.read_excel('data_karyawan.xlsx')
    print("Data yang dibaca dari file Excel:")
    display(df_excel)
except Exception as e:
    print(f"Error: {e}")
    print("Catatan: Jika terjadi error, pastikan library openpyxl sudah terinstall dengan menjalankan 'pip install openpyxl'")
```

### 2.3 Mengumpulkan Data dari API (Simulasi)

```python
# Simulasi pengambilan data dari API
import json

# Data JSON (simulasi respons API)
json_data = '''
{
  "data": [
    {"id": 1, "nama": "Fani", "usia": 27, "kota": "Jakarta"},
    {"id": 2, "nama": "Gita", "usia": 31, "kota": "Bandung"},
    {"id": 3, "nama": "Hadi", "usia": 24, "kota": "Surabaya"},
    {"id": 4, "nama": "Indra", "usia": 29, "kota": "Yogyakarta"},
    {"id": 5, "nama": "Joko", "usia": 33, "kota": "Semarang"}
  ]
}
'''

# Parse JSON
data_json = json.loads(json_data)

# Konversi ke DataFrame
df_api = pd.DataFrame(data_json['data'])

print("Data dari API (simulasi):")
display(df_api)
```

## 3. Visualisasi Data Berdasarkan Jenisnya

### 3.1 Visualisasi Data Numerik

```python
# Visualisasi data numerik
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Histogram untuk Jumlah Anak
sns.histplot(data_numerik['Jumlah Anak'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Histogram Jumlah Anak')
axes[0, 0].set_xlabel('Jumlah Anak')
axes[0, 0].set_ylabel('Frekuensi')

# Histogram untuk Tinggi Badan
sns.histplot(data_numerik['Tinggi Badan (cm)'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Histogram Tinggi Badan')
axes[0, 1].set_xlabel('Tinggi Badan (cm)')
axes[0, 1].set_ylabel('Frekuensi')

# Box plot untuk Jumlah Anak
sns.boxplot(y=data_numerik['Jumlah Anak'], ax=axes[1, 0])
axes[1, 0].set_title('Box Plot Jumlah Anak')
axes[1, 0].set_ylabel('Jumlah Anak')

# Box plot untuk Tinggi Badan
sns.boxplot(y=data_numerik['Tinggi Badan (cm)'], ax=axes[1, 1])
axes[1, 1].set_title('Box Plot Tinggi Badan')
axes[1, 1].set_ylabel('Tinggi Badan (cm)')

plt.tight_layout()
plt.show()
```

### 3.2 Visualisasi Data Kategorikal

```python
# Visualisasi data kategorikal
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Bar plot untuk Jenis Kelamin
sns.countplot(x='Jenis Kelamin', data=data_kategorikal, ax=axes[0])
axes[0].set_title('Distribusi Jenis Kelamin')
axes[0].set_xlabel('Jenis Kelamin')
axes[0].set_ylabel('Jumlah')

# Bar plot untuk Pendidikan
# Mengurutkan data ordinal
pendidikan_order = ['SMP', 'SMA', 'S1', 'S2', 'S3']
sns.countplot(x='Pendidikan', data=data_kategorikal, ax=axes[1], order=pendidikan_order)
axes[1].set_title('Distribusi Tingkat Pendidikan')
axes[1].set_xlabel('Tingkat Pendidikan')
axes[1].set_ylabel('Jumlah')

plt.tight_layout()
plt.show()
```

### 3.3 Visualisasi Hubungan Antar Data

```python
# Menggabungkan data untuk analisis
data_gabungan = pd.DataFrame({
    'Jumlah Anak': data_numerik['Jumlah Anak'],
    'Tinggi Badan': data_numerik['Tinggi Badan (cm)'],
    'Jenis Kelamin': data_kategorikal['Jenis Kelamin'],
    'Pendidikan': data_kategorikal['Pendidikan']
})

# Visualisasi hubungan antar data
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Scatter plot dengan warna berdasarkan jenis kelamin
sns.scatterplot(x='Tinggi Badan', y='Jumlah Anak', hue='Jenis Kelamin', data=data_gabungan, ax=axes[0])
axes[0].set_title('Hubungan Tinggi Badan dan Jumlah Anak berdasarkan Jenis Kelamin')
axes[0].set_xlabel('Tinggi Badan (cm)')
axes[0].set_ylabel('Jumlah Anak')

# Box plot tinggi badan berdasarkan pendidikan
sns.boxplot(x='Pendidikan', y='Tinggi Badan', data=data_gabungan, order=pendidikan_order, ax=axes[1])
axes[1].set_title('Distribusi Tinggi Badan berdasarkan Tingkat Pendidikan')
axes[1].set_xlabel('Tingkat Pendidikan')
axes[1].set_ylabel('Tinggi Badan (cm)')

plt.tight_layout()
plt.show()
```
